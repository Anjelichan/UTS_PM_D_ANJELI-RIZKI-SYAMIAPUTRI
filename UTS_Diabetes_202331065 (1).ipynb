# **preprocessing data**

```python
from sklearn.preprocessing import LabelEncoder
```

```python
#mengubah calues kolom yang bertipe string/object menjadi numerik

for col in ['gender','smoking_history']:
  label_encoder = LabelEncoder()
  dataset[col] = label_encoder.fit_transform(dataset[col])
```

```python
dataset #menampilkan dataset yang sudah diubah menjadi numerik
```

```python
x = dataset.drop(['diabetes'],axis=1)
y = dataset["diabetes"]
```

x= dataset.drop(['diabetes'],axis=1) = menghapus kolom diabetes dari tadaset dan menyimpanya dalam x, yang akan berisi semua fitur.

y= dataset["diabetes"] : menyimpan kolom diabetes sebagai target variabel y yang ingin diprediksi.

x berisi fitur untuk model dan y berisi target untuk prediksi.

# **splitting data**

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size=0.50)
```

membagi dataset menjadi 2 bagian yaitu satu untuk pelatihan dan sagtunya untuk pengujian. size 0.50 berarti 50% data pengujian. data pelatihan disimpan dalam x_train dan y_train, sedangkan data pengujian disimpan dalam x_test dan y_test

# **modelling data**

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(x_train, y_train)
```

gaussianNB membuat model klasifikasi menggunakan naive bayes dengan distribusi gausian. dan melatih model dengan data pelatihan x_train (fitur) dan y_train(target).

```python
hasil_prediksi = model.predict(x_test)
```

memprediksi label (target) untuk pengujian x_test

```python
from sklearn.metrics import accuracy_score
print('akurasi model:', accuracy_score(y_test,hasil_prediksi))
```

menghitung akurasi model dengan membandingkan prediksi hasil_prediksi dengan label asli y_test
