# **splitting data**

```python
x_train = dataset['Daily Exercise Time (X)'].values[:, np.newaxis]
y_train = dataset['Weight Loss (Y)'].values
```

kode yang mempersiapkan x_train 2D dengan data dari kolom daily exercise time(x) berbentuk (n_samples,1) sebagai variabel input  dan y_train  1D dengan data dari kolom weight loss (Y) bentuk (n_samples,) sebagai variabel tabel.

```python
x_train
```

# **modelling**

```python
from sklearn.linear_model import LinearRegression
```

linear_model adalah modul dalam sklearn yang berisikan tentang regresi dan klasifikasi berbasis regresi linear

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

kode diatas merupakan langkah dasar melatih model regresi linear dan setelah dilatih model siap untuk di prediksi

```python
prediksi = model.predict([[2]])
```

model yang telah dilatih untuk memprediksi nilai target (penurunan berat badan) berdasarkan input 2 ( model yang sudah dilatih)

```python
prediksi
```

menampilkan prediksi

# **testing data**

```python
x_train_test= [[3],[4.5]] #data-data baru
pred_y_train = model.predict(x_train_test)

print("Weight Loss : ",pred_y_train)
```

data baru (waktu latihan 3 dan 4.5) dimasukkan dalam model yang telah dilatih. model akan memprediksi penurunan berat badan untuk setiap data dan hasil akan ditampilkan dengan print().

```python
print("koefisien : ",model.coef_)
print("Intercept : ",model.intercept_)
```

pada model.coef_ menampilkan koefisien (slope) dari model regresi linear
menunjukkan seberapa besar pengaruh perubahan suatu nilai.
pada model.intercept_ menampilkan titik potong dengan sumbu y dari model. intercept adalah nilai prediksi y menunjukkan dimana garis dapat memotong sumbu y.

```python
y= model.coef_*27 + model.intercept_
```

menghitung prediksi y berdasarkan inputan 27 menggunakan koefisien dan intercept dari model regresi linear yang sudah dilatih.

```python
print(y)
```
